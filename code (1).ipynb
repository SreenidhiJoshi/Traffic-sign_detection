```python
import os

test_image_dir = "dataset/test/images"
images = os.listdir(test_image_dir)

print("Number of test images:", len(images))
print("First 5 images:", images[:5])
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

trained_model = YOLO("runs/detect/train12/weights/best.pt")

img_name = images[0]  # pick first image
img_path = f"dataset/test/images/{img_name}"

img = cv2.imread(img_path)

results = trained_model(img)

annotated_img = results[0].plot(labels=True, conf=True)
annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(annotated_img)
plt.axis("off")
```

```python
output_path = "output_labeled.jpg"
cv2.imwrite(output_path, cv2.cvtColor(annotated_img, cv2.COLOR_RGB2BGR))
print("Saved:", output_path)
```

```python
def estimate_distance(box_width, img_width):
    # Normalize width
    relative_size = box_width / img_width

    # Simple inverse relation (tuned values)
    if relative_size > 0.5:
        return 5
    elif relative_size > 0.3:
        return 10
    elif relative_size > 0.2:
        return 20
    elif relative_size > 0.1:
        return 50
    else:
        return 100
```
